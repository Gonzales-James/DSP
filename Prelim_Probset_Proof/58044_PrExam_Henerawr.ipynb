```python
def left_eq(x):
  ans = np.conj(x)
  return ans
```

```python
x = 1+3j
left_eq(x)
```

```python
def right_eq(x):
  ans = x ** -1
  return ans
```

```python
x = 2
right_eq(x)
```

```python
left_eq(x) == right_eq(x)
```

```python
def x_n(theta, n):
  eq = ( ( e**(1j*(pi/2)) + e**(-1j*(pi/2)) ) /(2)) - ( (( e**(1j * theta * n) - e**(-1j * theta * n) )) /(1j))
  return eq

def x_nsimp(theta, n):
  eq = ((e ** (1j * theta * n)) + (e ** (1j * theta * n))) * (1j)
  return eq

def plot(a, b):
  plt.plot(a, label="Original")
  plt.plot(b, label="Simplified")
  plt.legend()
  return plt.show()
```

```python
n = np.linspace(-5, 5, 50)
theta = 90

plot(x_n(theta, n), x_nsimp(theta, n))
```

# **4**

```python
def mov_average(l):
  i = 0
  new_l = []
  while len(l)-1 > i:
    average = round(np.sum(l[i:i+2]))/2
    new_l. append (average)
    i +=1
  return new_l
```

```python
arr = [1,2,3]
mov_average(arr)
```

```python
def avg_plot(l, l1):
  plt.plot(l, label="y[n]")
  plt.plot(l1, label="x[n]")
  plt.legend()
  return plt.show()

avg_plot(mov_average(arr), arr)
```

# **5**

```python
def y_n(arr,a,b,c):
  y_n = b * (a*arr + (arr-1)) - ((arr-3)*(c*arr + (arr-1)))
  return y_n
```

```python
arr = np.random.rand(10,)
a = np.linspace(-5, 5, 10)
b = np.linspace(-5, 5, 10)
c = np.linspace(-5, 5, 10)
```

```python
def plot(l, l1):
  plt.plot(l, label="x[n]")
  plt.plot(l1, label="y[n]")
  plt.legend()
  return plt.show()

plot(arr, y_n(arr, a,b,c))
```
