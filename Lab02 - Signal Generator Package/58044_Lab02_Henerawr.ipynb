##### Firstly, you need to upload the *siggenpy.py* file in the colab so that you can import it as a package without having any errors. After uploading the file you can now import it here in the colab. 

```python
import siggenpy as sig
```

##### After you have succesfully imported the package, you need to initialize the module *discrete_time* and enter the necessary parameters which are the minimum and maximum value for the x-axis.

```python
D_T = sig.discrete_time(-7, 7) # min, max are required parameters
```

### Impulse Signal

##### After the initialization of the module *discrete_time* is done, you can proceed in using the function *generate_signal* which is mainly used for generating a visualization or graph of the signal you desire. *generate_signal* function has only 2 required parameters, signal and the alpha that is required for some signals. Signals that are accepted as input parameters are impulse, unit-step, ramp, decay, growth, leaky, elu, and sigmoid. If other terms or keywords are inputted for the signal parameter, it will return an error.






```python
signal = "impulse"
D_T.generate_signal(signal)
```

### Unit Step

```python
signal = "unit-step"
D_T.generate_signal(signal)
```

### Ramp Signal

```python
signal = "ramp"
D_T.generate_signal(signal)
```

### Exponential Decay

##### In this case, decay signal is being utilized which means that the alpha parameter is required. If alpha parameter is not initialized, it will not return an error but instead it will utilized the alpha value that was initialized inside the package which is 0.01. This rule is also applicable to the other functions/signals that requires alpha parameter.

```python
signal = "decay"
D_T.generate_signal(signal, 0.5)
```

### Exponential Growth Signals

```python
signal = "growth"
D_T.generate_signal(signal, 2)
```

### Leaky ReLU Signal

```python
signal = "leaky"
D_T.generate_signal(signal, 0.1)
```

### ELU Signal

```python
signal = "elu"
D_T.generate_signal(signal, -0.5) # alpha should be inputted in negative
```

### Sigmoid Signal

```python
signal = "sigmoid"
D_T.generate_signal(signal)
```
